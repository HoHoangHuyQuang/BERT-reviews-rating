```python

from selenium.common.exceptions import *
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import os
```

```python
def write_file(data):
    df = pd.DataFrame(data, columns=['Comment', 'Label'])
   
    if os.path.exists('./data.xlsx'):
        base_file = pd.read_excel('./data.xlsx')   
        base_file = base_file.append(df, ignore_index= True)        
        base_file = base_file.loc[:, ~base_file.columns.str.contains('^Unnamed')]
        base_file = base_file.drop_duplicates(subset='Comment')
        base_file.to_excel('./data.xlsx')        
    else:        
        df.to_excel(r'E:\KLTN\data.xlsx')
```

```python
def load_url_selenium_tiki(url, filter = None):
    driver=  webdriver.Chrome(executable_path=r'E:\backup\chromedriver')
    print("Loading url=", url)
    driver.get(url)
    driver.maximize_window()
    list_review = []
    
    time.sleep(1.5)      
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
    if isinstance(filter, int):    
        try:    
            driver.find_element_by_xpath("//*[@class ='filter-review__text' and contains(text(), '"+ str(filter) +"')]").click()
            
        except:
            print('No filter')
            driver.close()
            return     
    while True:       
       
        try:          
            #Get the review details here
            WebDriverWait(driver, 5).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR,"div.review-comment__user")))            
            
        except :
            print('Not has comment!')
            break
        product_reviews = driver.find_elements_by_css_selector("div.review-comment__content")            
        for comment in product_reviews:                        
            if comment.text !='':
                if comment.find_elements_by_css_selector("span.show-more-content"):
                    comment.find_element_by_css_selector("span.show-more-content").click()
                if isinstance(filter, int):               
                    list_review.append([comment.text.strip(), filter-1])
                else:
                    list_review.append(comment.text.strip())
                            
        #Check for button next-pagination-item have disable attribute then jump from loop else click on the next button
        try:
            # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # time.sleep(1)
            pagination =driver.find_element_by_xpath("//div[@class='customer-reviews__pagination']")
            button_next= pagination.find_element_by_xpath("//a[@class='btn next']")
            driver.execute_script("arguments[0].click();", button_next)
            # print("next page")
            time.sleep(2)
           
        except (TimeoutException, WebDriverException) as e:
            print('Load several page!')
            break
    driver.close()
    return list_review
#tiki
```

```python
def load_url_selenium_shopee(url):
    driver=  webdriver.Chrome(executable_path=r'E:\backup\chromedriver')
    print("Loading url=", url)
    driver.get(url)
    driver.maximize_window()
    list_review = []
    
    
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight*2);")
    time.sleep(3)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.find_element_by_xpath("//div[contains(text(), '1 Sao')]").click()
    time.sleep(2)
    
    try:
        WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.CLASS_NAME,'product-rating-overview')))
        
    except:
        return 
    while True:
               
        try:          
            #Get the review details here
            WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR,"div.shopee-product-rating__main")))
           
        except :
            print('Not has comment!')
            break
        product_reviews = driver.find_elements_by_css_selector("div.shopee-product-rating")
        #            
        for comment in product_reviews:
            txt = comment.find_element_by_css_selector("div._3NrdYc")             
            if txt.text !='':                               
                list_review.append([txt.text.strip(), 1])
        #Check for button next-pagination-item have disable attribute then jump from loop else click on the next button
        try:
            
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            pagination = driver.find_element_by_xpath("//div[@class='shopee-page-controller product-ratings__page-controller']")
            button_next= pagination.find_element_by_class_name('shopee-svg-icon icon-arrow-right')
            # 
            driver.execute_script("arguments[0].click();", button_next)
            # print("next page")
            time.sleep(2)
            
        except (TimeoutException, WebDriverException) as e:
            print('Load several page!')
            break
    driver.close()
    return list_review
#shopee
```

```python
def load_url_selenium_fpt(url):
    driver=  webdriver.Chrome(executable_path=r'E:\backup\chromedriver')
    print("Loading url=", url)
    driver.get(url)
    driver.maximize_window()
    list_review = []
    
    time.sleep(1.5)
    #scroll to load
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight*2);")
    time.sleep(1)
    while True:        
        
        try:          
            #Get the review details here
            WebDriverWait(driver, 5).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR,"div.c-comment-box")))
        except :
            print('Not has comment!')
            break
        buyed_c = driver.find_element_by_css_selector("div.l-pd-rate")
        product_reviews = buyed_c.find_elements_by_css_selector("div.c-comment-text")
            
        for comment, rated in product_reviews:
            if comment.text !='':
                        
                list_review.append([comment.text.strip()])
        #Check for button next-pagination-item have disable attribute then jump from loop else click on the next button
        try:
            
            pagination =driver.find_element_by_xpath("//div[@class='c-comment-pagination margin-bottom-10 commentdetail']")
            button_next= pagination.find_element_by_xpath("//li[@title='Tiến 1 trang']")
            driver.execute_script("arguments[0].click();", button_next)
            # print("next page")
            time.sleep(2)
            
        except (TimeoutException, WebDriverException) as e:
            print('Load several page!')
            break
    driver.close()
    return list_review
#fpt
```
